# Q. 11) What is the Mean value of each column against each 'Weather Conditon' ?

```python
data.head(2)
```

```python
numeric_columns = ['Temp_C', 'Dew Point Temp_C', 'Rel Hum_%', 'Wind Speed_km/h', 'Visibility_km', 'Press_kPa']  # List the names of your numeric columns
mean_values = data.groupby('Weather Condition')[numeric_columns].mean()
```

```python
print(mean_values)
```
